```python
# get image from link google image

# first Image
urllib.request.urlretrieve(
'https://4.bp.blogspot.com/-DWHENNzZQEE/Vll8QUGGF9I/AAAAAAAAAKs/hRmfVHH57rY/s1600/2015-02-Aurora-Borealis-35-HD-Screensavers.jpg.', # link from google image
"gambar_1_logical.png")

# Second Image
urllib.request.urlretrieve(
'https://cdn.idntimes.com/content-images/community/2017/05/night-sky-sahale-glacierngsversion1495037976461adapt6761-aa35f472afd263ca8d858321435399c9.jpg', # link from google image
"gambar_2_logical.png")
```

```python
# load image
image1_logical = cv2.imread('gambar_1_logical.png')
image2_logical = cv2.imread('gambar_2_logical.png')

# Dapatkan resolusi gambar dalam piksel
width, height,channel = image1_logical.shape
print(f"Resolusi gambar 1: {1920}x{1200}x{channel} piksel")

# Dapatkan resolusi gambar dalam piksel
width_2, height_2,channel_2 = image1_logical.shape
print(f"Resolusi gambar 2: {1920}x{1200}x{channel_2} piksel")
```

```python
# plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(image1_logical, cmap='gray')
plt.title("Gambar satu")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image2_logical, cmap='gray')
plt.title("Gambar dua")
plt.axis('off')

plt.show()
```

or

```python
# hasil gambar Addition
result_logical = cv2.bitwise_or(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

and

```python
# hasil gambar Addition
result_logical = cv2.bitwise_and(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

NOT

```python
# hasil gambar Addition
result_logical = cv2.bitwise_not(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```
