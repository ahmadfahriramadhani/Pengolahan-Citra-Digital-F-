```python
# library
import urllib.request
import cv2 # https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html
import matplotlib.pyplot as plt
import numpy as np
```

```python
# first Image
urllib.request.urlretrieve(
'https://4.bp.blogspot.com/-DWHENNzZQEE/Vll8QUGGF9I/AAAAAAAAAKs/hRmfVHH57rY/s1600/2015-02-Aurora-Borealis-35-HD-Screensavers.jpg.', # link from google image
"gambar_1.png")

# Second Image
urllib.request.urlretrieve(
'https://cdn.idntimes.com/content-images/community/2017/05/night-sky-sahale-glacierngsversion1495037976461adapt6761-aa35f472afd263ca8d858321435399c9.jpg', # link from google image
"gambar_2.png")
```

```python
# load image
image_1 = cv2.imread('gambar_1.png') # cv2.IMREAD_GRAYSCALE
image_2 = cv2.imread('gambar_2.png')

# Resize image_2 to match the dimensions of image_1
image_2 = cv2.resize(image_2, (image_1.shape[1], image_1.shape[0]))
```

```python
# plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(image_1)
plt.title("Gambar satu")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_2)
plt.title("Gambar dua")
plt.axis('off')

plt.show()
```

```python
# hasil gambar Addition
result_add = cv2.add(image_1, image_2)
plt.imshow(result_add)
plt.title("Hasil Pengabungan")
plt.show()

# Dapatkan resolusi gambar dalam piksel
width, height, channel = result_add.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

```python
#--- Assuming you want to perform a thresholding operation for the logical images ---
# Adjust the threshold value (127 in this example) as needed
_, image1_logical = cv2.threshold(image_1, 127, 255, cv2.THRESH_BINARY)
_, image2_logical = cv2.threshold(image_2, 127, 255, cv2.THRESH_BINARY)

# Perform a logical operation (e.g., AND) on the thresholded images
result_logical = cv2.bitwise_and(image1_logical, image2_logical)

# Membandingkan data piksel
print("Image 1 Shape: ", image1_logical.shape)
print("Image 2 Shape: ", image2_logical.shape)
print("Result Image Shape: ", result_logical.shape)

# Jika ingin melihat beberapa piksel untuk perbandingan, ambil subset data
image1_pixels_logical = image1_logical[:5,:5]  # Ambil 5x5 bagian atas dari gambar 1
image2_pixels_logical = image2_logical[:5,:5]  # Ambil 5x5 bagian atas dari gambar 2
image3_pixels_logical = result_logical[:5, :5]  # Ambil 5x5 bagian atas dari gambar 2 #channel blue


print("\nData Piksel Gambar satu \n", image1_pixels_logical)
print("\nData Piksel Gambar dua \n", image2_pixels_logical)
print("\nData Piksel Gambar tiga \n", image3_pixels_logical)
```

```python
# load image
image1_logical = cv2.imread('gambar_1_logical.png')
image2_logical = cv2.imread('gambar_2_logical.png')


# Resize image_2 to match the dimensions of image_1
# image_2 = cv2.resize(image_2, (image_1.shape[1], image_1.shape[0]))
```

```python
# plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(image1_logical, cmap='gray')
plt.title("Gambar satu")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image2_logical, cmap='gray')
plt.title("Gambar dua")
plt.axis('off')

plt.show()
```

or

```python
# Resize both images to a common size before performing bitwise operations
image1_logical = cv2.resize(image1_logical, (1600, 1000))  # Or any other desired dimensions
image2_logical = cv2.resize(image2_logical, (1600, 1000))  # Ensure both images have the same size

```

```python
# hasil gambar Addition
result_logical = cv2.bitwise_or(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

and

```python
# hasil gambar Addition
result_logical = cv2.bitwise_and(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

NOT

```python
# hasil gambar Addition
result_logical = cv2.bitwise_not(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

XOR

```python
# hasil gambar Addition
result_logical = cv2.bitwise_xor(image1_logical, image2_logical,mask = None)
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```
