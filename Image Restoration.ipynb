```python
import cv2
import numpy as np
from scipy.stats.kde import gaussian_kde
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # Importing cv2_imshow from google.colab.patches
```

```python
# original image
f = cv2.imread('lenna.png', 0)
f = f/255
```

```python
plt.imshow(f, cmap='gray')
plt.show()
```

```python
# create gaussian noise
x, y = f.shape
mean = 0
var = 0.01
sigma = np.sqrt(var)
n = np.random.normal(loc=mean,
                     scale=sigma,
                     size=(x,y))
```

```python
plt.imshow(n, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# display the probability density function (pdf)
kde = gaussian_kde(n.reshape(int(x*y)))
dist_space = np.linspace(np.min(n), np.max(n), 100)
plt.plot(dist_space, kde(dist_space))
plt.xlabel('Noise pixel value'); plt.ylabel('Frequency')
plt.show()
```

```python
# add a gaussian noise
g = f + n
```

```python
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# original image
plt.subplot(1, 3, 1)
plt.imshow(f, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Gaussian noise
plt.subplot(1, 3, 2)
plt.imshow(n, cmap='gray')
plt.title('Gaussian Noise')
plt.axis('off')

# corrupted image
plt.subplot(1, 3, 3)
plt.imshow(g, cmap='gray')
plt.title('Corrupted Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
