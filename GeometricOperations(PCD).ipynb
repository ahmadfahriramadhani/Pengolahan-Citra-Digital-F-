```python
# library
import urllib.request
import cv2 # https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html
import matplotlib.pyplot as plt
import numpy as np
```

```python
# first Image
urllib.request.urlretrieve(
'https://i.pinimg.com/736x/20/2e/aa/202eaa724f6982f2b8a40a0c1058286a.jpg', # link from google image
"gambar_1.png")
```

```python
# load image
image_1 = cv2.imread('gambar_1.png')

# Convert BGR image to RGB
image_rgb = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB) # Changed 'image' to 'image_1'
```

```python
# plot image
plt.imshow(image_rgb)
plt.axis('off') # Turn off axis
plt.title('yuta normal')
plt.show()
```

```python
# prompt: rotatet image

(h, w) = image_rgb.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(image_rgb, M, (w, h))
plt.imshow(rotated)
plt.axis('off')
plt.title('yuta rotated')
plt.show()
```

```python
#reflection image
reflected = cv2.flip(image_rgb, 1)  # 1 for horizontal flip
plt.imshow(reflected)
plt.axis('off')
plt.title('yuta reflected')
plt.show()
```

```python
# translation image

# Translation
tx = 50  # Shift along x-axis
ty = 100 # Shift along y-axis

M = np.float32([[1, 0, tx], [0, 1, ty]])
translated = cv2.warpAffine(image_rgb, M, (w, h))

plt.imshow(translated)
plt.axis('off')
plt.title('yuta translated')
plt.show()
```

```python
# combine translation,rotate, & reflection image
# Combine Transformations

# Rotate the image
(h, w) = image_rgb.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(image_rgb, M, (w, h))

# Translate the rotated image
tx = 50
ty = 100
M = np.float32([[1, 0, tx], [0, 1, ty]])
translated_rotated = cv2.warpAffine(rotated, M, (w, h))


# Reflect the translated and rotated image
reflected_translated_rotated = cv2.flip(translated_rotated, 0)

# Display the combined transformed image
plt.imshow(reflected_translated_rotated)
plt.axis('off')
plt.title('Combined Transformation')
plt.show()
```
